```python
# Leemos otros datos.

data=pd.read_csv("https://raw.githubusercontent.com/Albertuff/Machine-Learning/master/datos/Telco-Customer-Churn.csv")
data.head()
```

```python
# Dimensión de los datos
data.shape
```

```python
# Que tipo de datos
data.info()
```

```python
# Detectamos que TotalCharges es de tipo cadena cuando deberia ser de tipo numerico
# Transformamos la columa a tipo de datos numericos con ayuda de to_numeric de pandas
cargos_totales=pd.to_numeric(data.TotalCharges,errors="coerce") # error="coerce" obliga a que los datos que no son numeros los codifique como NaN (Not A Number)
cargos_totales
```

```python
# Hay datos NaN
cargos_totales.isna()
# Para saber cuantos datos faltan se suman
cargos_totales.isna().sum()
```

```python
# Para hacer la sustitucion directamente en el dataframe
data.TotalCharges=pd.to_numeric(data.TotalCharges,errors="coerce")
data.info()
```

```python
# Total charges es la unica variable que tiene datos faltantes
data.isna().sum()
```

```python
# Vamos a sustituir los datos faltantes por cero 
data.TotalCharges=data.TotalCharges.fillna(0)
data.isna().sum()
```

```python
# Descripcion de los datos
data.describe()
```

```python
# Para manipular el dataframe se estandarizan los nombres de los atributos y las respuestas en los atributos que son cadenas (tipo object)
# Cambiamos los n ombres de las columnas a minusculas

data.columns=data.columns.str.lower().str.replace(" ","_")
# str.lower(): convierte los nombres a minusculas
# str.replace(): cambia los espacios en blanco (" ") por un guión bajo ("_")
data.columns
```

```python
# Conectar las respuestas en atributos que son de tipo cadena. Identificamos primero los atributos que son de tipo object
atributos_str=list(data.dtypes[data.dtypes=="object"].index)
atributos_str
for col in atributos_str:
    data[col]=data[col].str.lower().str.replace(" ","_")
data.head()
```

```python
# Lo usual es que las cualidades, principalmente las de la variable de respuesta sean numéricas
data.churn=(data.churn=="yes").astype(int)      # True:1   &  False:0
data.gender=(data.gender=="female").astype(int)     # True:1  &  False:0
data.head()
```

```python
data.describe()
```
